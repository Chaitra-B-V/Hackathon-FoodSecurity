```python
tas1 = pd.read_csv(
    'Data/Climatology/tas_climatology_annual-monthly_cru_1961-1990_IND.csv')
tas2 = pd.read_csv(
    'Data/Climatology/tas_climatology_annual-monthly_cru_1991-2020_IND.csv')
together = [tas1, tas2]
tas = pd.concat(together)
tas = tas.loc[tas["States"] == 'Uttar Pradesh']
tasdata = {
    '1961-1990': tas.iloc[0][1:14].values,
    '1991-2020': tas.iloc[1][1:14].values
}
tasdf = pd.DataFrame(tasdata,
                        index=[
                            'Annual', 'Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun',
                            'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'
                        ])
tasdf.plot()
plt.show()
```
